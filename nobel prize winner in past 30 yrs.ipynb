```python
# import packages
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
```

```python
df1 = pd.DataFrame(columns=['name','title','description','Born','Spouse','Thesis','Education','Notable students','Awards','Parents','Discovered','Books','Kown for','Nationality','Full Name','Academic advisor','Movies','year'])
```

```python
driver = webdriver.Chrome('/Applications/chromedriver')
```

```python
for yr in range(2017,2020):
    # open driver and website
    driver.get('https://www.google.com/')
    sleep(0.5)
    driver.find_element_by_name('q').send_keys('nobel prize winners ' + str(yr))
    sleep(0.5)
    driver.find_element_by_class_name('gNO89b').click()
    url = driver.current_url
    
    df = pd.DataFrame(columns=['name','title','description','Born','Spouse','Thesis','Education','Notable students','Awards','Parents','Discovered','Books','Kown for','Nationality','Full Name','Academic advisor','Movies'])
    numpeople = len(driver.find_elements_by_xpath('//div[@class="kIXOkb PZPZlf"]'))
    
    for p in range(numpeople):
        driver.find_elements_by_xpath('//div[@class="kIXOkb PZPZlf"]')[p].click()
        
        name = driver.find_element_by_xpath('//div[@class="SPZz6b"]/h2/span').get_attribute('innerHTML')
        try:
            title = driver.find_element_by_xpath('//div[@class="wwUB2c PZPZlf"]/span').get_attribute('innerHTML')
        except:
            pass
        description = driver.find_element_by_xpath('//div[@class="kno-rdesc"]/div/span').get_attribute('innerHTML')
        df.loc[p] = ''
        df['name'][p] = name
        df['title'][p] = title
        df['description'][p] = description
        
        num = len(driver.find_elements_by_xpath('//span[@class="GRkHZd w8qArf"]'))
    
        for i in range(num):
            name = ''
            cont = ''
            namepath = '//div[@class="mod"]/following-sibling::div['+ str(i+1) +']/div/div/div/span[@class="GRkHZd w8qArf"]/a'
            name = driver.find_element_by_xpath(namepath).get_attribute('innerHTML')
            contentpath = '//div[@class="mod"]/following-sibling::div['+ str(i+1) +']/div/div/div/span[@class="Eq0J8 LrzXr kno-fv"]/a'
            try:
                content = driver.find_element_by_xpath(contentpath).get_attribute('innerHTML')
            except:
                pass
            cont = content
            j = 0
            while j <= 4:
                j = j + 1
                if content[0] != '<':
                    try:
                        contentpath = '//div[@class="mod"]/following-sibling::div['+ str(i+1) +']/div/div/div/span[@class="Eq0J8 LrzXr kno-fv"]/a/following-sibling::a['+ str(j) +']'
                        content = driver.find_element_by_xpath(contentpath).get_attribute('innerHTML')
                        if content[0] != '<':
                            cont = cont + '; ' + content
                    except:
                        pass
            try:
                df[name][p]=cont
            except:
                pass

    df['year'] = str(yr)
    
    df1 = pd.concat([df1,df])

df1
```
