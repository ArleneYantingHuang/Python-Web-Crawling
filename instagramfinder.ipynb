```python
from urllib.request import urlopen
from bs4 import BeautifulSoup #BeautifulSoup 
url="https://blog.feedspot.com/tech_instagram_influencers/"
content=urlopen(url)
soup=BeautifulSoup(content,'html.parser') #BeautifulSoup 
print(soup)
```

```python

for i in range(0,99):
    name = soup.select("#fsb")[0].select("p")[2*i].text.strip()[3:1000]
    bio = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[0].next_sibling.strip()
    insurl = soup.select("#fsb")[0].select("p")[(1+2*i)].select('a')[0].get('href').strip()
    followers = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[2].next_sibling.strip()
    l = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")
    location = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[(len(l)-1)].next_sibling.strip()
    website = soup.select("#fsb")[0].select("p")[(1+2*i)].select("a")[1].get_text().strip()
```

```python
import pandas as pd

df = pd.DataFrame(columns=['name','bio','insurl','followers','location','website'])

for i in range(0,100):
    name = soup.select("#fsb")[0].select("p")[2*i].text.strip()[3:1000]
    bio = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[0].next_sibling.strip()
    insurl = soup.select("#fsb")[0].select("p")[(1+2*i)].select('a')[0].get('href').strip()
    followers = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[2].next_sibling.strip()
    l = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")
    location = soup.select("#fsb")[0].select("p")[(1+2*i)].select("strong")[(len(l)-1)].next_sibling.strip()
    if l == 5:
        website = soup.select("#fsb")[0].select("p")[(1+2*i)].select("a")[1].get_text().strip()
    row = [name,bio,insurl,followers,location,website]
    df.loc[i] = row

```
