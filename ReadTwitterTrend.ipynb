```python
import json
with open('json.json') as f:
    data = json.load(f)
```

```python
import pandas as pd
df1 = pd.DataFrame(columns=['Top Hashtags (San Fransisco)','Tweet_volume'],index=list(range(1,50)))
```

```python
for i,line in enumerate(data[0]['trends']):
    Hashtag = line['name'].replace('#','')
    Tweet_volume = line['tweet_volume']
    row = [Hashtag,Tweet_volume]
    df1.loc[i+1] = row
```
