```python
from urllib.request import urlopen
from bs4 import BeautifulSoup #BeautifulSoup 
url="https://blog.feedspot.com/technology_youtube_channels/"
content=urlopen(url)
soup=BeautifulSoup(content,'html.parser') #BeautifulSoup 
print(soup)
```

```python
rank = soup.findAll("div", class_="fsb v4")
```

```python
soup.select("#fsb > p")[0].img.get('alt')
```

```python
import pandas as pd

df = pd.DataFrame(columns=['name','location','url','about','frequency','since'])
data = []

for i in range(0,74):
    name = soup.select("#fsb > p")[i].img.get('alt')
    location = soup.select("#fsb > p")[i].select("span")[1].get_text()
    url = soup.select("#fsb > p")[i].select("a")[0].get("href")
    about = soup.select("#fsb > p")[i].select('strong')[0].next_sibling.strip()
    frequency = soup.select("#fsb > p")[i].select('strong')[1].next_sibling.strip()
    since = soup.select("#fsb > p")[i].select('strong')[2].next_sibling.strip()
    data = [name,location,url,about,frequency,since]
    df.loc[i] = data
```
