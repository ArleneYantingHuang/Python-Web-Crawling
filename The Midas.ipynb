```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib
```

```python
url = 'https://www.forbes.com/profile/neil-shen/?list=midas&sh=6c4bc10e21ff'
content=urlopen(url)
soup=BeautifulSoup(content,'html.parser')
print(soup)
```

```python
df = pd.DataFrame(columns=['name','title','about','otherlist','Age','Source of Wealth','Residence',
                           'Citizenship','Marital Status','Education','Notable Deal'])
```

```python
name = soup.title.get_text().strip()
title = soup.find_all('div', class_ = 'profile-subheading')[0].get_text().strip()
about = soup.find_all('div', class_ = 'profile-text')[0].get_text().strip()
otherlist = []
for line in soup.find_all('div', class_ = 'profile-list__item'):
    rank = line.select('.profile-list__rank')[0].get_text().strip()
    award = line.select('a')[0].get_text().strip()
    onlist = rank + ' on forbes list of ' + award
    otherlist.append(onlist)
row = [name,title,about,otherlist,'','','','','','','']
df.loc[0] = row
for line in soup.find_all('div', class_ = 'profile-stats__item'):
    statname = line.select('.profile-stats__title')[0].contents[0]
    stattext = line.select('.profile-stats__text')[0].contents[0]
    df[statname][0] = stattext

    
    
    
```
